```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import datetime as dt
```

```python
# Create a Ticker object
apple_ticker = yf.Ticker("AAPL")
apple_historical_data = apple_ticker.history(period='max')
exp_dates = apple_ticker.options
apple_options = pd.DataFrame()
for exp_date in exp_dates:
    feat_data = apple_ticker.option_chain(exp_date).calls.dropna()
    feat_data = feat_data[['lastTradeDate', 'strike', 'lastPrice', 'bid', 'ask', 'impliedVolatility']]
    feat_data['expirationDate'] = exp_date
    apple_options = pd.concat([apple_options, feat_data])
```

```python
apple_options['DaysToExpiration'] = (pd.to_datetime(apple_options['expirationDate']) - pd.Timestamp.now()).dt.days
```

```python
# Clean data

aapl_close = apple_historical_data['Close'].dropna()
aapl_returns = aapl_close.diff()[1:]/aapl_close[1:]*100
apple_options.dropna(inplace=True)
```
